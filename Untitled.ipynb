```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
carbon_records=pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-18/food_consumption.csv')
carbon_records.head()
```

```python
carbon_records.info()
```

```python
east_african=carbon_records.loc[carbon_records['country'].isin(['Kenya','Uganda','Tanzania','Rwanda','Ethiopia'])]
east_african.head()
```

```python
mercy=east_african.groupby('food_category')['co2_emmission'].mean()
print(mercy)
```

```python
mercy.sort_values(ascending=False).plot.bar()
plt.xlabel('food_category')
plt.ylabel('co2_emmission')
plt.title('co2_emmission by food_category')
plt.show()
```

```python
mercymerine=pd.DataFrame(mercy.reset_index())
```

```python
mercymerine.sort_values(by='co2_emmission',ascending=False)
```

```python
plt.barh(mercymerine['food_category'],mercymerine['co2_emmission'])
plt.show()
```

```python
sns.barplot(mercymerine,y='food_category',x='co2_emmission',orient='h')
plt.show()
```

```python

sns.boxplot(y='food_category',x='co2_emmission',data=mercymerine)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.catplot(x='co2_emmission', y ='food_category', data =mercymerine, kind='box', hue='food_category' )
plt.xlabel('Co2 Emission(Kg CO2/person/year)')
plt.ylabel('Food_Category')
plt.title('Co2 Emission')
plt.grid(True)
plt.show()
```
