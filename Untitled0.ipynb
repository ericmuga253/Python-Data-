```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sport=pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-04-07/tdf_winners.csv')
sport.head()
```

```python
sport.info()
```

```python
missing_sports=(sport.isnull().sum()/len(sport))*100
missing_sports
```

```python
 sport.drop(columns=missing_sports[missing_sports>40].index,inplace=True,axis=1)
```

```python
mean_time=sport['time_overall'].mean()
sport['time_overall'].fillna(mean_time,inplace=True)
```

```python
mean_time_margin=sport['time_margin'].mean()
sport['time_margin'].fillna(mean_time,inplace=True)
```
