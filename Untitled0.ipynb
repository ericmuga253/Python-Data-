```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sport=pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-04-07/tdf_winners.csv')
sport.head()
```

```python
sport.info()
```

```python
missing_sports=(sport.isnull().sum()/len(sport))*100
missing_sports
```

```python
 sport.drop(columns=missing_sports[missing_sports>40].index,inplace=True,axis=1)
```

```python
sns.lineplot(data=sport, x="edition", y="time_overall",)
```

```python
sport['time_overall'] = pd.to_timedelta(sport['time_overall'], unit='h')
sport['time_margin'] = pd.to_timedelta(sport['time_margin'], unit='h')

# Calculate the runner_up_time
sport['runner_up_time'] = sport['time_overall'] + sport['time_margin']

# Display the DataFrame to verify the new column
print(sport[['edition', 'time_overall', 'time_margin', 'runner_up_time']])
```

```python
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# Line plot for Winning Time
sns.lineplot(data=sport, x='edition', y='time_overall', label='Winning Time', marker='o')

# Line plot for Time Margin
sns.lineplot(data=sport, x='edition', y='time_margin', label='Time Gap (Runner-Up)', marker='o')

plt.xlabel('Edition')
plt.ylabel('Time (hours)')
plt.title('Winning Time and Time Gap (Runner-Up) per Edition')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
