```python
df= pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-04-21/gdpr_violations.tsv', sep='\t')
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
df = df.sort_values(by='date')

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['price'], marker='o', linestyle='-', color='b')

# Adding title and labels
plt.title('Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
df.tail()
```

```python
start_date = '2018-01-01'
end_date = '2020-12-31'
mask = (df['date'] >= start_date) & (df['date'] <= end_date)
df_filtered = df.loc[mask]

# Sort the filtered DataFrame by the 'date' column
df_filtered = df_filtered.sort_values(by='date')

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df_filtered['date'], df_filtered['price'], marker='o', linestyle='-', color='b')

# Adding title and labels
plt.title('Price from 2018 to 2020')
plt.xlabel('Date')
plt.ylabel('Price')
```

```python
df['year'] = df['date'].dt.year
```

```python
df = df[(df['date'].dt.year >= 2018) & (df['date'].dt.year <= 2020)]

# Determine the top 10 articles violated based on total price
top_10_articles = df.groupby('article_violated')['price'].sum().nlargest(20).index

# Filter DataFrame to include only data related to top 10 articles violated
df_top_10_articles = df[df['article_violated'].isin(top_10_articles)]

# Group and sum the data by year and article violated
grouped_data = df_top_10_articles.groupby(['year', 'article_violated'])['price'].sum().unstack(fill_value=0)

# Plotting the data
grouped_data.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='tab20')

plt.xlabel('Year')
plt.ylabel('Total Price')
plt.title('Total Price by Year and Top 20 Articles Violated')
plt.xticks(rotation=0)
plt.legend(title='Article Violated', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
max_price = df['price'].max()
print("The maximum price in the DataFrame is:", max_price)
```

```python
max_price_rows = df[df['price'] == max_price]

print("The maximum price in the DataFrame is:", max_price)
print("Row(s) with the maximum price:")
print(max_price_rows)
```

```python
grouped_by_country = df.groupby('name')['price'].sum().reset_index()

# Print the maximum price to determine the scale
max_price = grouped_by_country['price'].max()
print("The maximum price paid for violations is:", max_price)

# Sort the data by price for better visualization
grouped_by_country = grouped_by_country.sort_values(by='price', ascending=True)

# Plotting the data
plt.figure(figsize=(12, 8))
plt.barh(grouped_by_country['name'], grouped_by_country['price'], color='skyblue')
plt.xlabel('Total Price Paid for Violations')
plt.ylabel('country')
plt.title('Total Price Paid for Violations by Country')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
grouped_by_company = df.groupby('controller')['price'].sum().reset_index()

# Sort the data by price in descending order and select the top 10 companies
top_10_companies = grouped_by_company.sort_values(by='price', ascending=False).head(10)

# Print the maximum price to determine the scale
max_price = top_10_companies['price'].max()
print("The maximum price paid for violations among the top 10 companies is:", max_price)

# Plotting the data
plt.figure(figsize=(12, 8))
plt.barh(top_10_companies['controller'], top_10_companies['price'], color='skyblue')
plt.xlabel('Total Price Paid for Violations')
plt.ylabel('Company')
plt.title('Top 10 Companies by Total Price Paid for Violations')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest price at the top
plt.tight_layout()

# Display the plot
plt.show()
```

```python
drop=['pop_est','continent','iso_a3','gdp_md_est','id','picture','authority','date','controller','type']
europe.drop(drop, axis=1, inplace=True)
europe.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
europe.boundary.plot(ax=ax)
europe.plot(column='price', ax=ax, legend=True,
            legend_kwds={'label': "price",
                         'orientation': "horizontal"},
            cmap='plasma', missing_kwds={'color': 'lightgrey'})

# Adding titles and labels
ax.set_title('Total Fines Paid in European Countries', fontdict={'fontsize': 20})
ax.set_axis_off()

# Show the plot
plt.show()
```

```python
ax.legend(title='Article Violated', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10, title_fontsize=12)

# Annotate bars with values
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    if height > 0:  # Only annotate bars with height greater than 0
        ax.annotate(f'{height:,.0f}', (x + width / 2, y + height / 2),
                    ha='center', va='center', fontsize=8, color='black', rotation=90)

# Tight layout for better spacing
plt.tight_layout()
```
